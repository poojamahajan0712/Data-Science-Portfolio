```python
from transformers import BartTokenizer, BartForConditionalGeneration

 # Model
checkpoint = 'facebook/bart-large-xsum'
# Loading Tokenizer
tokenizer = BartTokenizer.from_pretrained(checkpoint)
# Loading Model
model = BartForConditionalGeneration.from_pretrained(checkpoint)
```

Loading News Summarisation data

```python
import pandas as pd
import re
df = pd.read_excel("/content/drive/MyDrive/AI Projects/NewsSummarisation/datasets/Inshorts Cleaned Data.xlsx")
```

```python
article = df['Short']
summary = df['Headline']
```

```python
def preprocess(text):
    text = re.sub(r"&.[1-9]+;"," ",text)
    return text
article = article.apply(lambda x: preprocess(x))
summary = summary.apply(lambda x: preprocess(x))
```

```python
def generate_summary(text):
    inputs = tokenizer(text, max_length=1024, return_tensors="pt", truncation=True)
    # Generate summary (tune max_length and min_length for better summaries)
    summary_ids = model.generate(inputs["input_ids"], max_length=60, min_length=40, length_penalty=2.0, num_beams=4, early_stopping=True)
    # Decode and print summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    print(summary)
```
