```python
def cal_year_month_cnt_graph(data,col):
    merged_calc = data.groupby(['yr','mnth'])[col].sum().reset_index()
    merged_calc['year_mon'] = merged_calc['yr'].astype(str)+"_"+merged_calc['mnth'].astype(str)
    merged_calc.drop(['yr','mnth'],axis=1,inplace=True)

    plt.figure(figsize=(10, 5))
    plt.bar(merged_calc['year_mon'],merged_calc[col])
    
```

```python
def draw_pie_graph(df, grp_col,agg_col):
    temp_df= df.groupby(grp_col)[agg_col].sum().reset_index()
    temp_df[agg_col] = (temp_df[agg_col]/temp_df[agg_col].sum())*100
    plt.pie(temp_df[agg_col],labels = temp_df[grp_col].astype(str),autopct='%1.1f%%')
    plt.show()
    
```

### Section 3 - Model building 

```python
main_data.dtypes
```

```python
main_data.drop(['date1'],axis=1,inplace=True)
```

```python
## creating dummy columns
cat_col_dummies = pd.get_dummies(main_data.drop(['dteday'],axis=1),drop_first=True)

cat_col_dummies.dtypes
```

```python
y = main_data['cnt']
X = cat_col_dummies.drop(['instant','casual','registered','cnt','atemp'],axis=1)
```

```python
X.head()
```

```python
## splitting between train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
def draw_res(model,y_train,X_train):
    pred_act = pd.DataFrame({'Actual': y_train.values,'Predicted':model.predict(X_train)})
    pred_act['residual'] = pred_act['Actual'] - pred_act['Predicted']
    pred_act['residual'].hist()
```

```python
def draw_res_fitted(model,y_train,X_train):
    pred_act = pd.DataFrame({'Actual': y_train.values,'Predicted':model.predict(X_train)})
    pred_act['residual'] = pred_act['Actual'] - pred_act['Predicted']
    
    plt.figure(figsize=(8, 6))
    plt.scatter(pred_act['Predicted'], pred_act['residual'], color='blue', edgecolor='k')
    plt.axhline(y=0, color='red', linestyle='--', linewidth=2)  # Horizontal reference line at y=0
    plt.xlabel('Fitted Values')
    plt.ylabel('Residual')

    
```

Linear Regression

```python
import statsmodels.api as sm
new_X_train = sm.add_constant(X_train)

# Fit the model
reg_model = sm.OLS(y_train, new_X_train).fit()

# Get the summary of the model, including p-values
print(reg_model.summary())
```

```python
#In the context of linear regression,
#the null hypothesis is that the coefficient of a particular predictor is equal to zero (i.e., the predictor has no effect on the dependent variable).
#P-value ≤ 0.05: The predictor is considered statistically significant. There is strong evidence to reject the null hypothesis, meaning the predictor likely has an effect on the dependent variable.

# important variables based on p values 
## temp, humidity, windspeed, season, year, workingday, weathersit
```

```python
new_data = sm.add_constant(X_test)
draw_res(reg_model,y_test,new_data)
## error range from -3000 to 1000
```

```python
print(sqrt(mean_squared_error(y_train,reg_model.predict(new_X_train))))
print(sqrt(mean_squared_error(y_test,reg_model.predict(new_data))))
## quite high and consistent in both sets of data, it means values will be off by 750 rentals quite high.
# for an avergae of 4.5k rentals daily values can be off by 17% approx
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(reg_model.fittedvalues, reg_model.resid, color='blue', edgecolor='k')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)  # Horizontal reference line at y=0
plt.xlabel('Fitted Values')
plt.ylabel('Residual')

## resiudals symmetric around 0, pattern observed close to higher fitted values,large outliers also present,
#will try non linear models also 
```

```python
plt.scatter(y_train.values,reg_model.predict(new_X_train))
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r--', lw=2) 

## wider spread around line, indicating large residuals
## shows clear trend though, moderately accurate
```

2. RandomForestRegressor

```python
## this is after some trials , using GridSearchCV with RandomForest regressor and using rmse to measure performance 
params = {'n_estimators': [200,300,400,500],
           'max_depth': [10,15,20,30],
           'max_features' : ['auto','log2']}
    
rfmodel1 = RandomForestRegressor(
#                                  min_samples_split=15,
#                                  min_samples_leaf=8,
#                                  min_impurity_decrease=0.02,
                                 random_state =118)
                                 #oob_score=True)
gsearch1 =GridSearchCV(estimator = rfmodel1, param_grid = params, scoring='neg_mean_squared_error', cv=7)

gsearch1.fit(X_train,y_train)
```

```python
best_model = gsearch1.best_estimator_
best_model
```

```python
print(sqrt(mean_squared_error(y_train,best_model.predict(X_train))))
print(sqrt(mean_squared_error(y_test,best_model.predict(X_test))))

## 714 error means the prediction will be off by 714 , this can further be improved, however for a naive modle prediction i.e average 4.5k , this number is better.
```

```python
draw_res(best_model,y_test,X_test) ## looks pretty much following normal distribution
```

```python
feat_imp = pd.DataFrame( {'col':list(X_train.columns),'feature_imp':best_model.feature_importances_})

plt.figure(figsize=(20, 6))
plt.barh(feat_imp['col'],feat_imp['feature_imp'])
## temp, year, hum,windspeed, season2  comes to be good predictors
```

```python
plt.scatter(y_train.values,best_model.predict(X_train))
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r--', lw=2) 

# The model appears to perform well across most of the range, with fitted values closely matching actual values for much of the data.
# A few points at the lower end could be potential outliers.
## better fit that linear regression, as line is tighly fit 
```

```python
best_model.score(X_test,y_test) ## r2 value - 85%
```

```python
draw_res_fitted(best_model,y_train,X_train)
```

3. Using XgboostRegressor 

```python
# Instantiation 
xgb_r = xgb.XGBRegressor(objective ='reg:squarederror', n_estimators = 10, seed = 123) 
# Fitting the model 
xgb_r.fit(X_train,y_train) 
```

```python
print(sqrt(mean_squared_error(y_train,xgb_r.predict(X_train))))
print(sqrt(mean_squared_error(y_test,xgb_r.predict(X_test))))
```

```python
## tuning hyperparameters

params = {'learning_rate': [0.01,0.5,0.1],
            'n_estimators': [100,200,500,600],
            'max_depth': [5,10,12,15]}

xgb1 = xgb.XGBRegressor(objective ='reg:squarederror',seed=1234)

gsearch_xgb =GridSearchCV(estimator = xgb1, param_grid = params, scoring='neg_mean_squared_error', cv=7)

gsearch_xgb.fit(X_train,y_train)
```

```python
best_model_xgb = gsearch_xgb.best_estimator_
best_model_xgb

```

```python
print(sqrt(mean_squared_error(y_train,best_model_xgb.predict(X_train))))
print(sqrt(mean_squared_error(y_test,best_model_xgb.predict(X_test))))

## better than random forest , and vanilla xgboost.
```

```python
best_model_xgb.score(X_test,y_test) # r2 for xgboost model 
```

```python
plt.scatter(y_train.values,best_model_xgb.predict(X_train))
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r--', lw=2) 
## more tighter fit around line, even lower end residuals, outliers are least, 
```

```python
draw_res(best_model_xgb,y_test,X_test) ## range of errors reduced , although looks normally distributed.
```

```python
draw_res_fitted(best_model_xgb,y_train,X_train)
```

```python
# Get feature importance
importance = best_model_xgb.feature_importances_

# Plotting feature importance
plt.barh(X_train.columns, importance)
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.show()
# year, season 4, temp and weathersit3(fall) comes to be good predictors
```
